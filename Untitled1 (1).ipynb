```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
from keras.preprocessing import image
from sklearn.utils.class_weight import compute_class_weight
import os
from tensorflow.keras.applications import MobileNetV2
```

```python
dataset_path = "/content/drive/MyDrive/Lung_Cancer_Dataset"
```

```python
# Image parameters
IMG_SIZE = (224, 224)
BATCH_SIZE = 32
```

```python
# Data Augmentation & Preprocessing
datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    zoom_range=0.3,
    horizontal_flip=True,
    width_shift_range=0.2,
    height_shift_range=0.2,
    validation_split=0.2
)

train_data = datagen.flow_from_directory(
    dataset_path,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training'
)

val_data = datagen.flow_from_directory(
    dataset_path,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation'
)
```

```python
# Get class names
class_names = list(train_data.class_indices.keys())
```

```python
# Compute Class Weights to Handle Imbalance
labels = train_data.classes
class_weights = compute_class_weight(
    class_weight="balanced",
    classes=np.unique(labels),
    y=labels
)
class_weights = dict(enumerate(class_weights))
```

```python
# Load Pretrained MobileNetV2 Model (Fine-Tuned)
base_model = keras.applications.MobileNetV2(
    weights="imagenet",
    input_shape=(224, 224, 3),
    include_top=False
)
```

```python
# Unfreeze last few layers for better feature extraction
for layer in base_model.layers[-30:]:
    layer.trainable = True
```

```python
# Custom Classification Layers
model = keras.Sequential([
    base_model,
    keras.layers.GlobalAveragePooling2D(),
    keras.layers.Dense(256, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(3, activation='softmax')
])
```

```python
# Compile Model (Optimized for Accuracy)
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.00005),  # Lower LR for fine-tuning
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
# Train Model with Class Weights
history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=15,
    class_weight=class_weights,  # Balances the classes
    verbose=1
)
```

```python
# Function to Predict User Uploaded Image
def predict_ct_scan():
    uploaded = files.upload()
    for filename in uploaded.keys():
        img_path = filename
        img = image.load_img(img_path, target_size=(224, 224))
        img_array = image.img_to_array(img) / 255.0
        img_array = np.expand_dims(img_array, axis=0)

        prediction = model.predict(img_array)
        predicted_class = class_names[np.argmax(prediction)]

        # Display Image and Prediction
        plt.imshow(img)
        plt.axis('off')
        plt.title(f"Predicted Case: {predicted_class}")
        plt.show()

# Call Function to Upload and Predict
predict_ct_scan()
```
